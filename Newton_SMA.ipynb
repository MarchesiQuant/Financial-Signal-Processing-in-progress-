```python
# METODO DE NEWTON PARA LA wc de la SMA
from math import sin, cos, sqrt, pi
from scipy.optimize import newton
import matplotlib.pyplot as plt 
import numpy as np
plt.style.use('seaborn')
```

```python
def get_sma_cutoff(N, **kwargs):
    func = lambda w: sin(N*w/2) - N/sqrt(2) * sin(w/2)  # |H(e^jω)| = √2/2
    deriv = lambda w: cos(N*w/2) * N/2 - N/sqrt(2) * cos(w/2) / 2  # dfunc/dx
    omega_0 = pi/N  # Starting condition: halfway the first period of sin(Nω/2)
    return newton(func, omega_0, deriv, **kwargs)
```

```python
wc = 0.1
f1 = lambda N: np.sin(wc*N/2)**2
f2 = lambda N: (N**2/2)*(np.sin(wc/2)**2)

N = np.linspace(1,100)

y1 = f1(N)
y2 = f2(N)
func = y1-y2 
df = lambda N: (wc/2)*np.cos(wc*N/2) - (1/np.sqrt(2))*np.sin(wc/2)
y3 = df(N)

# Create the plot with labels, titles, and custom line styles/colors
# plt.plot(N, y1, label='f1(N)', color='blue')
# plt.plot(N, y2, label='f2(N)', color='red')
plt.plot(N,y3)

# Add labels and a title
plt.xlabel('N')
plt.ylabel('Value')

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.6)
fig = plt.gcf()
fig.set_size_inches(6, 4) 
plt.show()
```

```python
# N DE LA SMA 
def N_sma(wc, **kwargs):
    func = lambda N: np.sin(wc*N/2) - (N/np.sqrt(2))*(np.sin(wc/2))
    deriv = lambda N: (wc/2)*np.cos(wc*N/2) - (1/np.sqrt(2))*np.sin(wc/2)
    N_0 = pi/wc  
    return newton(func, N_0, deriv, **kwargs)
```

```python
# WC DE LA SMA 
def wc_sma(N, **kwargs):
    func = lambda w: sin(N*w/2) - N/sqrt(2) * sin(w/2)  # |H(e^jω)| = √2/2
    deriv = lambda w: cos(N*w/2) * N/2 - N/sqrt(2) * cos(w/2) / 2  # dfunc/dx
    omega_0 = pi/N  # Starting condition: halfway the first period of sin(Nω/2)
    return newton(func, omega_0, deriv, **kwargs)
```
